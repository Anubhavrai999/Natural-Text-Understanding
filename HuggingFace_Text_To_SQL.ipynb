# Text To SQL Query Helper Tool using Hugging Face

````python
!pip install -q transformers accelerate torch
````

````python
from transformers import pipeline
````

## Load Hugging Face Model

````python
pipe = pipeline(
    "text-generation",
    model="TinyLlama/TinyLlama-1.1B-Chat-v1.0"
)
````

````python
from langchain_community.llms import HuggingFacePipeline
llm = HuggingFacePipeline(pipeline = pipe, model_kwargs = {'temperature':0})
````

## Create Prompt

````python
template = """
             Create a SQL query snippet using the below text:
              ```{text}```
              Just SQL query:
           """
````

````python
from langchain_core.prompts import PromptTemplate

prompt = PromptTemplate(template=template, input_variables=["text"])




````

````python
llm_chain = prompt | llm
````

````python
text = """ Extract all the unique values from column "age"
"""
````

````python
print(llm_chain.invoke(text)[0])
````

````python
print(f"Type of pipe: {type(pipe)}")
print(f"Type of llm.pipeline: {type(llm.pipeline)}")
print(f"Is llm.pipeline the same object as pipe? {llm.pipeline is pipe}")
````

If `llm.pipeline is pipe` returns `False`, it confirms the `llm` object was initialized incorrectly. The `llm.pipeline` should be of type `transformers.pipelines.text_generation.TextGenerationPipeline` and be the same object as `pipe`.
